```python
exp_lvl = {'EN':'Entry-level',
            'MI':'Mid-level',
            'SE':'Senior-level',
            'EX':'Executive-level'}
ds_salary['experience_level'] = ds_salary['experience_level'].map(exp_lvl)
```

```python
emp_type = {'PT':'Part-time',
            'FT':'Full-time',
            'CT':'Contract',
            'FL':'Freelance'}
ds_salary['employment_type'] = ds_salary['employment_type'].map(emp_type)
```

# **Univariate**

```python
import matplotlib.pyplot as plt

# Assuming 'salary_in_usd' is a column in your DataFrame
plt.hist(ds_salary['salary_in_usd'], bins=20, color='skyblue', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
print("Mean salary:", ds_salary['salary_in_usd'].mean())
print("Median salary:", ds_salary['salary_in_usd'].median())
print("Standard deviation:", ds_salary['salary_in_usd'].std())
```

```python
sns.violinplot(x='salary_in_usd', data=ds_salary)
plt.title('Violin Plot of Salary')
plt.show()
```

```python
sns.countplot(x='experience_level', data=ds_salary)
plt.title('Count of Experience Levels')
plt.show()
```

# **Bivariate**

```python
result1 = ds_salary.groupby('experience_level')['salary_in_usd'].agg(['mean', 'min', 'max'])
result1['mean'] = result1['mean'].round(2)

# Resetting index to make 'experience_level' a column for Seaborn plotting
result1.reset_index(inplace=True)

# Creating subplots for three different bar plots
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plotting mean salary
sns.barplot(ax=axes[0], x='experience_level', y='mean', data=result1, color='skyblue')
axes[0].set_title('Mean Salary by Experience Level')
axes[0].set_xlabel('Experience Level')
axes[0].set_ylabel('Salary (Mean)')

# Plotting max salary
sns.barplot(ax=axes[1], x='experience_level', y='max', data=result1, color='salmon')
axes[1].set_title('Max Salary by Experience Level')
axes[1].set_xlabel('Experience Level')
axes[1].set_ylabel('Salary (Max)')

# Plotting min salary
sns.barplot(ax=axes[2], x='experience_level', y='min', data=result1, color='lightgreen')
axes[2].set_title('Min Salary by Experience Level')
axes[2].set_xlabel('Experience Level')
axes[2].set_ylabel('Salary (Min)')

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
# Assuming 'ds_salary' is your DataFrame
df_2023 = ds_salary[ds_salary['work_year'] == 2023]
top_countries = pd.DataFrame(df_2023['company_location'].value_counts(normalize=True).head())
total_top_countries = top_countries.sum()
top_countries.loc['Others'] = 1 - total_top_countries

ax = top_countries.plot(kind='bar', legend=False, color='skyblue')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax.set_ylabel('% of total jobs')
ax.set_xlabel('Company Location')
ax.set_title('Distribution of Jobs by Location in 2023', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='experience_level', y='salary_in_usd', data=ds_salary, color='purple', alpha=0.7)
plt.title('Scatter Plot of Property Size vs. Price')
plt.xlabel('experience_level')
plt.ylabel('salary_in_usd')
plt.show()
```

# **Multivariate**

```python
import seaborn as sns

# Choose specific variables for the pair plot
pairplot_variables = ['work_year', 'experience_level', 'salary_in_usd']

# Create a pair plot
sns.pairplot(ds_salary[pairplot_variables], hue='experience_level', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Work Year, Experience Level, and Salary', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a 1x2 grid for subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Subplot 1: Salary Comparison Across Experience Levels and Employment Types
sns.barplot(x='experience_level', y='salary_in_usd', data=ds_salary, hue='employment_type', ci=None, palette='Set2', ax=axes[0])
axes[0].set_title('Salary Comparison Across Experience Levels and Employment Types')
axes[0].set_xlabel('Experience Level')
axes[0].set_ylabel('Salary (USD)')
axes[0].legend(title='Employment Type')

# Subplot 2: Salary Distribution Over Years
sns.boxplot(x='work_year', y='salary_in_usd', data=ds_salary, palette='coolwarm', ax=axes[1])
axes[1].set_title('Salary Distribution Over Years')
axes[1].set_xlabel('Work Year')
axes[1].set_ylabel('Salary (USD)')

# Adjust layout
plt.tight_layout()
plt.show()
```
